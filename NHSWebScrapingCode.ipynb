# Extract NHS condition names and their url

```python
url = "https://www.nhsinform.scot/illnesses-and-conditions/a-to-z"
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
text = soup.get_text()
all_disease = soup.find_all("a", {"class": "col small-12 medium-6 large-4 module"})
list_title_rows = []
list_href_col = [] 

for d in all_disease:
    sigle_data = []
    sigle_data.append(d.get("title"))
    sigle_data.append("https://www.nhsinform.scot"+d.get("href"))
    list_title_rows.append(sigle_data) 
df = pd.DataFrame(list_title_rows, columns=['condition','url'])
df
```
